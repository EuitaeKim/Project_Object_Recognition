```python
import pandas as pd
```

```python
# Review & Info 데이터 불러오기
url = '...dir'

df = pd.read_csv(url + '7. ulsan_store_review_nam.csv')
df.head()

# 리뷰는 정규식 적용
# 주소는 필요한 앞부분만 가져오기
```

```python
df['add_summary_1'] = df['소재지(도로명)']
df['add_summary_2'] = df['소재지(도로명)']

for idx, i in enumerate(df['add_summary_1']) :
    try :
        result = list(str(i).split())[1]
        df['add_summary_1'][idx] = result
    except :
      pass

for idx, i in enumerate(df['add_summary_2']) :
    try :
        result = list(str(i).split())[2]
        df['add_summary_2'][idx] = result
    except :
      pass

df = df[['add_summary_1', 'add_summary_2', 'ItemID', 'Timestamp', 'Rating', 'review']]
df.head()
```

```python
import re
# 전처리 진행
df['review_pre'] = df['review']

# 기본 전처리
text_desc = []
for text in df['review'] :
    text = text.replace('\n', ' ').strip()
    # doc_tokens = re.sub(r"[^a-zA-Z0-9ㄱ-ㅣ가-힣]", " ", str(text))
    doc_tokens = text.replace('  ', ' ').replace('   ', ' ').replace('    ', ' ').replace('     ', ' ')
    doc_tokens = doc_tokens.lower()
    text_desc.append(doc_tokens)

df.loc[:]['review_pre'] = text_desc
```

```python
df = df[['add_summary_1', 'add_summary_2', 'ItemID', 'Timestamp', 'Rating', 'review_pre']]
df.head()
```
