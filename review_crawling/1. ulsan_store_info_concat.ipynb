```python
# 필요한 칼럼만 추출
df = df[['업소명', '소재지(도로명)']]
df2 = df2[['업소명', '소재지(도로명)']]
df3 = df3[['업소명', '소재지']]
df4 = df4[['가맹점명', '소재지도로명주소']]
df3 = df3.rename(columns={'소재지':'소재지(도로명)'})
df4 = df4.rename(columns={'가맹점명':'업소명', '소재지도로명주소':'소재지(도로명)'})

df_fin = pd.concat([df, df2, df3, df4])
df_fin.reset_index(inplace=True, drop=True)
df_fin['naver_map_url'] = ''
```
